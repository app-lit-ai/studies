```python
import yfinance as yf
```

```python
ticker = yf.Ticker("TSLA")
```

```python
df = ticker.history(period="max")
```

if the price the previous 4 days or more is below the 10-day simple moving average for each of those days AND the current price above 10 day simple moving average then buy it now, it will pop sometime the next day

```python
df['ma_10'] = df.Close.rolling(window=10).mean()
df['HighNext'] = df.High.shift(-1)
```

```python
mask = (df.Close > df.ma_10) & (df.Close.shift(1) < df.ma_10) & (df.Close.shift(2) < df.ma_10) & (df.Close.shift(3) < df.ma_10) & (df.Close.shift(4) < df.ma_10)
condition = df.loc[mask]
condition
```

```python
wins, losses = sum(condition.HighNext > condition.Close), sum(condition.HighNext < condition.Close)
total = wins + losses
fpct = lambda x: '{:.1%}'.format(x)

wins, fpct(wins / total), losses, fpct(losses / total), total, 1
```
